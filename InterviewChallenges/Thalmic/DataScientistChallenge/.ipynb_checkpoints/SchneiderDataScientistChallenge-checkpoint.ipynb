Task: Classifiy an Individual based on their Gait from Accelerometer Data Collected by Their Phone

Welcome! I will be walking through the process of training machine learning algorithms to successfully re-identifying an individual based on the X, Y, and Z accelerometer data recorded from their phone. We will be using the 'User Identification From Walking Activity Data Set' which includes the sensor data for 22 individuals.

First thing we'll need to do is import the libraries we'll be using. We'll be using SKLearn for our basic classification algorithms, and Keras for our deep learning methods. I also set a seed for reproducibility.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import re
import statistics
import pickle
import random
from sklearn import neighbors, tree, svm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.utils import to_categorical
np.random.seed(3)
```

We get an deprecation error here involving one of the Keras package. We can ignore it. To get started we'll need to load the csv files and run some basic summary statistics. We'll be checking to make sure all data values are real numbers and if any of the data are empty. We'll then plot some basic statistics to get a sense of the variability in the data. 

```python
gaitfiles = os.listdir('WalkingActivity')
gaitdict = {}
gaitlenlist = []
nan, empty = False, False

for file in gaitfiles:
    gaitname = 'person' + file.split('.')[0]
    gaitdict[gaitname] = np.genfromtxt('WalkingActivity/{}'.format(file), delimiter=',')
    if np.isnan(gaitdict[gaitname]).any():
        nan = True
        print('{} has NAN data'.format(file))
    if gaitdict[gaitname].size == 0:
        empty = True
        print('{} is empty.'.format(file))
    gaitlenlist.append(len(gaitdict[gaitname]))

if nan == False: print('No NAN in data')
if empty == False: print('No Empty data files\n')

print('Min Time Entries: {}'.format(min(gaitlenlist)))
print('Max Time Entries: {}'.format(max(gaitlenlist)))
```

We have no NAN in our data and all our loaded files have data. So that's a great start. But Woah! There's a huge disparity between the number of time steps recorded for each individual. This is something we need to be aware of moving forward as our results could suffer from this class imbalance if we sample evenly from each individual. Let's take a closer look at the variability of the length of the data. 

```python
plt.boxplot(gaitlenlist)
plt.xlabel('Gait Profiles')
plt.ylabel('Number of time series entries')
plt.title('Variation of time series entries')
plt.show()

print('Average Time Steps: {}'.format(round(sum(gaitlenlist) / len(gaitlenlist))))
print('STD Dev Time Steps: {}'.format(round(statistics.stdev(gaitlenlist))))
```

Hmm... So it's not just one or two outliers. The standard deviation of 6051 shows we have a wide disparity across our data. Let's keep going for now however we may need to address this later if our models perform poorly. One tactic to address this would be to use some form of bootstrapping to create additional samples of the individuals with less amount of data. 

Let's now take a look at what our X, Y, Z data actually look like by plotting their values for the first 3 individuals.

```python
x1 = gaitdict['person1'][:, 0]
y1 = gaitdict['person1'][:, 1]
y2 = gaitdict['person1'][:, 2]
y3 = gaitdict['person1'][:, 3]
plt.plot(x1, y1, label='X Response')
plt.plot(x1, y2, label='Y Response')
plt.plot(x1, y3, label='Z Response')
plt.xlabel('Time Steps')
plt.ylabel('Signal Value')
plt.title('X, Y, and Z Signal Values for Individual 1')
plt.legend()
plt.show()

x1 = gaitdict['person2'][:, 0]
y1 = gaitdict['person2'][:, 1]
y2 = gaitdict['person2'][:, 2]
y3 = gaitdict['person2'][:, 3]
plt.plot(x1, y1, label='X Response')
plt.plot(x1, y2, label='Y Response')
plt.plot(x1, y3, label='Z Response')
plt.xlabel('Time Steps')
plt.ylabel('Signal Value')
plt.title('X, Y, and Z Signal Values for Individual 2')
plt.legend()
plt.show()

x1 = gaitdict['person3'][:, 0]
y1 = gaitdict['person3'][:, 1]
y2 = gaitdict['person3'][:, 2]
y3 = gaitdict['person3'][:, 3]
plt.plot(x1, y1, label='X Response')
plt.plot(x1, y2, label='Y Response')
plt.plot(x1, y3, label='Z Response')
plt.xlabel('Time Steps')
plt.ylabel('Signal Value')
plt.title('X, Y, and Z Signal Values for Individual 3')
plt.legend()
plt.show()
```

There are some interesting things to note here. The first is that there appears to be quite a large amount of variation within one individuals gait pattern. This is likely as a result of stopping and starting while walking. It will be curious to see if our model can accurately predict the individual even during these nuances. Second, it appears that the frequency of oscillation are unique for each individual which is promising for our models to be able to classify an individual's pattern. Let's plot the X sensor data for the 3 individuals to get a closer look at this and compare if the individual oscillations do in fact differ. 

```python
x1 = gaitdict['person1'][:, 0]
x2 = gaitdict['person2'][:, 0]
x3 = gaitdict['person3'][:, 0]
y1 = gaitdict['person1'][:, 1]
y2 = gaitdict['person2'][:, 1]
y3 = gaitdict['person3'][:, 1]
plt.plot(x1, y1, label='Individual 1')
plt.plot(x2, y2, label='Individual 2')
plt.plot(x3, y3, label='Individual 3')
plt.xlabel('Time Steps')
plt.ylabel('X Response')
plt.title('Comparison of X Response of 3 Individuals')
plt.legend()
plt.show()
```

Neat! From this we can see clearly that the X sensor data between individuals is quite different which should be promising for our models. We can also see the disparity in the length of the time series data between the individuals here.

So... How should we approach this problem? There is no specification for the requirements of how to ID a person, so let's set an ambitious objective for ourselves! What if we test to see if we can build a model to classifiy an individual based on their gait after just 1 second. That would be impressive and useful! Alright well, how many time steps are in 1 second? First we need to see if the time series responses are even across individuals. Let's take a look.

```python
onesecondlist = []
for i in range(len(gaitfiles)):
    onesecondlist.append((np.abs(gaitdict['person{}'.format(i+1)][:, 0] - 1)).argmin())

stepsonesecond = int(round(sum(onesecondlist) / len(onesecondlist)))

print('List of the number of time steps until 1 second:\n {}'.format(onesecondlist))
print('\nAverage number of time steps until 1 second: {}'.format(stepsonesecond))
```

Well. The indexs aren't identical across all individuals, but they're pretty close. Since they are so close, let's consider the average 32 time steps to equal 1 second for all of our samples. 

With that in mind, it's time to create our data. The simplest approach to get started is to ignore the time component and consider the one second interval as a single chunk of data. We can include time series models later if our initial models fail to perform. To transform the data, I flatten the 3x32 X, Y, Z responses into a 1x96 length feature vector that follow the pattern of X, Y, Z, X, Y, Z, X, Y, Z,... Then, for each individual we are going to take the flattened 1 second segments and randomly distribute them with a 9/10 chance into a training data set and 1/10 chance into testing data set with each 1 second interval assigned their correct classification. The randomness introduced in the train/test ensures that we are unbaised in our distribution of the data and we are testing our ability to indentify the individual in a variety of circumstances. This also allows us to perform statistics on the generalization of the models by changing the random seed and rerunning the models. 

```python
train_data = []
test_data = []

for key in gaitdict:
    lentimeseries = len(gaitdict[key])
    counter = 0
    individual = int(re.search(r'\d+', key).group()) - 1
    while True:
        if counter + stepsonesecond > lentimeseries: 
            break
        oneseconddata = gaitdict[key][counter:counter + stepsonesecond, 1:]
        oneseconddata = [item for sublist in oneseconddata for item in sublist]
        counter += stepsonesecond
        if np.random.random() > 0.1: train_data.append([oneseconddata, individual])
        else: test_data.append([oneseconddata, individual])

random.Random(1).shuffle(train_data)
random.Random(1).shuffle(test_data)

x_train = np.array([i[0] for i in train_data])
y_train = np.array([i[1] for i in train_data])

x_test = np.array([i[0] for i in test_data])
y_test = np.array([i[1] for i in test_data])

y_train_onehot = to_categorical(y_train)
y_test_onehot = to_categorical(y_test)

num_classifications = len(set(y_test))

unique, counts = np.unique(y_train, return_counts=True)
trainclassificationdict = dict(zip(unique, counts))

averagesamples = 0
for key in trainclassificationdict:
    averagesamples += trainclassificationdict[key]
averagesamples = averagesamples/num_classifications

print('Number of training set: \t', len(x_train))
print('Number of test set:\t\t', len(x_test))
print('Number of unique classifications: ', num_classifications)
print('\nAverage number of training examples for each individual: ', round(averagesamples))
```

Great! So we've created a modestly sized training set of 4211 examples and 447 test examples to see how well we do. Because the length of the data for each individual is different, the number of training examples per individual is not the same. They're instead proportional to the length of time series data given. Let's take a look at that. If our models are unsuccessful this could be a reason why. 

```python
plt.bar(trainclassificationdict.keys(), trainclassificationdict.values(), 1, color='g')
plt.xlabel('Individual')
plt.ylabel('Number of Training Examples')
plt.title('Training Set Distribution')
plt.show()
```

We can see here clearly that the number of training examples per individuals is disproportion. We'll revisit this later if we need to. Alright! Let the machine learning begin. What I do here is start simple and work my way up in complexity to build an ensemble of models. Worth noting before considering our results is that based on 22 classifications, a purely random baseline model would have approximately a 1/22 chance, or 4.5% accuracy, on the test set. 

The first model I'll use is K-Nearest Neighbours which will make a classification prediction based on the features of the K-closest data entries seen during training. For each model I print out the accuracy on the training set, the test set, as well as a diagnostic confusion matrix illustrating the distribution of correct and incorrect classifications. For the confusion matrix, the X axis represents classifications from individual 1 to 22 in the data set left to right and the Y axis represent the predicted classifications from individual 1 to 22 top to bottom. Numbers across the diagonal indicate the number of correct classifications. Numbers in the upper right indicate false negatives. Numbers in the  lower left indicate false positives. Typically in confusion matrices we're looking for a row or column that has a disproportional number of incorrect answers which is indicative of a classification the model is finding particularly difficult to correctly classify.

```python
if not os.path.exists('saved_models'):
    os.makedirs('saved_models')

nnclf = neighbors.KNeighborsClassifier(n_neighbors=5)
nnclf.fit(x_train, y_train)
trainaccuracy = nnclf.score(x_train, y_train)
testaccuracy = nnclf.score(x_test, y_test)
with open('saved_models/nnclf', 'wb') as f:
    pickle.dump(nnclf, f)

print('Train Accuracy: ', round(trainaccuracy, 3))
print('Test Accuracy: ', round(testaccuracy, 3))
print(confusion_matrix(y_test, nnclf.predict(x_test)))
```

K-Nearest Neighbours returned 69.4%! Considering we are dealing with 22 classifications this is actually quite good for our first model. Despite it's simplicity K-Nearest Neighbours is an extremely robust. Let's try Random Forest next which create an ensemble of logic based decision trees considering the training features and then each tree votes for the final classification.

```python
rfclf = tree.DecisionTreeClassifier()
rfclf.fit(x_train, y_train)
trainaccuracy = rfclf.score(x_train, y_train)
testaccuracy = rfclf.score(x_test, y_test)
with open('saved_models/rfclf', 'wb') as f:
    pickle.dump(rfclf, f)
    
print('Train Accuracy: ', round(trainaccuracy, 3))
print('Test Accuracy: ', round(testaccuracy, 3))
print(confusion_matrix(y_test, rfclf.predict(x_test)))
```

Typical for Random Forest we have massively overfit our training data with an accuracy of 100% and 44.5% on our testing data. Not great, but considering a baseline of 4.5%, it's still not terrible. If we count the from the left of the matrix, we can see individual 17 gave this model a hard time. Let's try a support vector machine next which tries to optimize the distance of the decision boundaries between classifications using a Gaussian kernel. 

```python
svmclf = svm.SVC(kernel='rbf', max_iter=-1)
svmclf.fit(x_train, y_train)
trainaccuracy = svmclf.score(x_train, y_train)
testaccuracy = svmclf.score(x_test, y_test)
with open('saved_models/svmclf', 'wb') as f:
    pickle.dump(svmclf, f)

print('Train Accuracy: ', round(trainaccuracy, 3))
print('Test Accuracy: ', round(testaccuracy, 3))
print(confusion_matrix(y_test, svmclf.predict(x_test)))
```

65.1% accuracy. Respectable. A little worse than K-Nearest Neighbour. Once again individual 17 causing problems. If this were a real problem at this point I would investigate this data set closer to see if there was something unique about it. Let's try next an One-Vs-All Logistic Regression which tries to create a linear boundary for each classification. We're dealing with a 96-dimensionally oscillation pattern so a linear decision boundary may might not end well...

```python
logitclf = LogisticRegression()
logitclf.fit(x_train, y_train)
trainaccuracy = logitclf.score(x_test, y_test)
testaccuracy = logitclf.score(x_train, y_train)
with open('saved_models/logitclf', 'wb') as f:
    pickle.dump(logitclf, f)

print('Train Accuracy: ', round(trainaccuracy, 3))
print('Test Accuracy: ', round(testaccuracy, 3))
print(confusion_matrix(y_test, logitclf.predict(x_test)))
```

36.8% accuracy... Not great. As expected, it's hard to create linear decision boundaries considering this data. It is still much better than the 4.5% baseline of random. 

Well... that is a collection of the applicable greatest hits of machine learning approaches used before 2012. Let's go one step further and see if we can train a neural network to do better than these models. First let's train a simple one layer network with 1028 hidden units to output the correct classification. We'll be using Keras here which is a readable and convienent API overlay for TensorFlow. As a method to improve generalization I also introduce a Dropout of 20% to the network. What Dropout does is during each iteration of training 20% of the nodes in the hidden layer are randomly inactive. This helps reduce overfitting. For the sake of demonstration I will only train 10 epochs here and load a trained model after 250 epochs and achieved an accuracy of 77.4% (the best yet!). If you'd like to edit the number of epochs to watch the accuracy continue to increase feel free to do so. 

```python
fnnmodel = Sequential()
fnnmodel.add(Dense(1028, activation='relu', name='fc1', kernel_initializer='glorot_normal', 
                   input_dim=len(x_train[0])))
fnnmodel.add(Dropout(0.2))
fnnmodel.add(Dense(len(set(y_test)), activation='softmax'))

fnnmodel.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

fnnmodel.fit(x_train, y_train_onehot, epochs=10, batch_size=32, validation_data=(x_test, y_test_onehot))

fnnmodel.load_weights('saved_models/FNN-0.774.hdf5')
```

Next, let's train a deep neural network. This means we'll train a network that has multiple layers which can map more complex logical relationships within the data. Once again, for demonstration purposes I only train for 10 epochs and load a model that was trained for 250 epochs. It acheived a slightly improved performance of 77.8% accuracy. Best yet!

```python
dnnmodel = Sequential()
dnnmodel.add(Dense(1028, activation='relu', name='fc1', kernel_initializer='glorot_normal', 
                   input_dim=len(x_train[0])))
dnnmodel.add(Dropout(0.2))
dnnmodel.add(Dense(1028, activation='relu', name='fc2', kernel_initializer='glorot_normal'))
dnnmodel.add(Dropout(0.2))
dnnmodel.add(Dense(len(set(y_test)), activation='softmax'))


dnnmodel.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

dnnmodel.fit(x_train, y_train_onehot, epochs=10, batch_size=32, validation_data=(x_test, y_test_onehot))

dnnmodel.load_weights('saved_models/DNN-0.778.hdf5')
```

So... 77.8% accuracy was our best model. Is that the end of the story? What is holding us back from achieving higher accuracy? Is 1 second too short? Due to the nature of our data do we have too large of a class imbalance? Should we provide augmented data to improve the performance of our neural networks? Maybe we should double our training data by adding 1 second data samples beginning at 0.5 seconds rather than 0 seconds? Should we consider a time series analysis? That would involve us having to reformat our data... Before any of these drastic steps, there's one more thing we can try. Since each machine learning technique has its own limitations, a powerful technique is known as an ensemble, where we consider the output of all of our models in combination, allowing for each to provide a vote on the correct output classification. Let's see what that looks like. 

Here we're going to create a list for each testing example of the output classifications of each model and consider our final prediction to be the classification which has the most votes. In the case where there is no replicated answer, the response of our deep neural network is used because it had the highest test accuracy. Also at this time I record the test data of incorrect classifications used for plotting later.

```python
correct = 0
incorrectdict = {}

for i in range(len(test_data)):
    predictlist = [int(nnclf.predict(x_test[i:i+1])),
                   int(rfclf.predict(x_test[i:i+1])),
                   int(svmclf.predict(x_test[i:i + 1])),
                   int(logitclf.predict(x_test[i:i + 1])),
                   np.argmax(fnnmodel.predict(np.array([x_test[i]]))),
                   np.argmax(dnnmodel.predict(np.array([x_test[i]])))]
    if len(predictlist) > len(set(predictlist)):
        answer = np.argmax(dnnmodel.predict(np.array([x_test[i]])))
    else: answer = max(set(predictlist), key=predictlist.count)

    if answer == y_test[i]: correct += 1
    else:
        if y_test[i] not in incorrectdict: incorrectdict[y_test[i]] = 1
        else:
            if 'incorrectanswers' not in locals(): incorrectanswers = x_test[i]
            else: incorrectanswers = np.vstack((incorrectanswers, x_test[i]))
            incorrectdict[y_test[i]] += 1

print('Ensemble Accuracy: ', round(correct/len(y_test), 3))

print('\nNumber of Incorrect Answers Per Individual:\n\n', incorrectdict)
```

Accuracy of... 94.0% Wahoo! For the purpose of this challenge I would consider this a success. In a real life scenario I would either report the results to management for next steps or refer to the documentation provided by the client as a reference to what their expectations are. We can see that the same individual 17 from the confusion matrices has the largest number of incorrect classifications. Outside of 17 there isn't an abnormal amount of incorrect classifications for a particular classification which allows us to relieve our concerns about class imbalance. Let's take a closer look at exactly what our ensemble model is predicting by plotting where our model gives its correct and incorrect results. In the code below I reshape each test data point back to it's original 3x32 shape and scan the original data for the same array to get the appropriate x co-ordinates. I then do the same for the test data labelled incorrectly by the model. I then plot each individual and highlight the test data correctly answered in green, and test data incorrectly answered in red. 

```python
counter = 0
for key in gaitdict:
    while counter < len(gaitdict[key]):
        for i in range(len(x_test)):
            if np.array_equal(gaitdict[key][counter:counter + stepsonesecond, 1:], x_test[i].reshape(32, 3)):
                if 'tested' not in locals():
                    tested = gaitdict[key][counter:counter + stepsonesecond]
                    tested = np.vstack((tested, np.array([np.nan, np.nan, np.nan, np.nan])))
                else:
                    tested = np.vstack((tested, gaitdict[key][counter:counter + stepsonesecond]))
                    tested = np.vstack((tested, np.array([np.nan, np.nan, np.nan, np.nan])))
        for i in range(len(incorrectanswers)):
            if np.array_equal(gaitdict[key][counter:counter + stepsonesecond, 1:], incorrectanswers[i].reshape(32, 3)):
                if 'wrong' not in locals():
                    wrong = gaitdict[key][counter:counter + stepsonesecond]
                    wrong = np.vstack((wrong, np.array([np.nan, np.nan, np.nan, np.nan])))
                else:
                    wrong = np.vstack((wrong, gaitdict[key][counter:counter + stepsonesecond]))
                    wrong = np.vstack((wrong, np.array([np.nan, np.nan, np.nan, np.nan])))
        counter += 32
    try:
        x1 = gaitdict[key][:, 0]
        y1 = gaitdict[key][:, 1]
        x2 = tested[:, 0]
        y2 = tested[:, 1]
        x3 = wrong[:, 0]
        y3 = wrong[:, 1]
        plt.plot(x1, y1, color='black', label='Training Data')
        plt.plot(x2, y2, color='green', label='Testing Data Correctly Classified')
        plt.plot(x3, y3, color='red', label='Testing Data Incorrectly Classified')
        plt.xlabel('Seconds')
        plt.ylabel('Signal Value')
        plt.title('Individual {} X Axis Gait Data'.format(int(re.search(r'\d+', key).group()) - 1))
        plt.legend()
        plt.show()
    except:
        x1 = gaitdict[key][:, 0]
        y1 = gaitdict[key][:, 1]
        x2 = tested[:, 0]
        y2 = tested[:, 1]
        plt.plot(x1, y1, color='black', label='Training Data')
        plt.plot(x2, y2, color='green', label='Testing Data Correctly Classified')
        plt.xlabel('Seconds')
        plt.ylabel('Signal Value')
        plt.title('Individual {} X Axis Gait Data'.format(int(re.search(r'\d+', key).group()) - 1))
        plt.legend()
        plt.show()
    counter = 0
    del tested
    if 'wrong' in locals(): del wrong
```

I find these graphs informative for a number of reasons. The first is that it shows overall how successful my classifier is, but also where exactly it failed to perform. By visually inspecting the results we can see that the vast majority of the 1 second periodic oscillations were classified correctly. We can also see from these plots that the majority of incorrect classifications occurred during periods of time when little to no oscillation occurred. What I find remarkable is that the model still performed quite well on a large number of testing examples that occurred during atypical oscillations. When looking at the troublesome individual 17, it certainly appears the incorrect classifications were during periods of time with little oscillation. When considering using this model in practice, by visually inspecting the data like this we can see that one reasonable technique we can use to improve the model's performance is to limit identification to periods of time when the accelerometer data is above a minimum threshold.

I consider 94% test accuracy to be a success for the purpose of this challenge but there are further techniques we could use try to improve performance. The first would be to double the data by gathering data starting at 0.5 seconds rather than 0 seconds to increase the amount of training data. This is especially important for neural networks. A second might be to introduce what is known as a polynomial feature expansion which increases the number of features by multiplying them together capturing novel interactions. A third would be to take advantage of the time dependent nature of the data and train either a hidden-markov model or Long Short-Term Memory Recurrent Neural Network.
